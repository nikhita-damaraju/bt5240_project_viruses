```python
#program to merge all csv file data into one single file
import os
import glob
import pandas as pd
import numpy as np
os.chdir("/Users/nikhitadamaraju/Desktop/Systems Biology/Project/RNA_negative_viruses")
all_filenames = glob.glob("*human.csv")

def unique(list1): 
    x = np.array(list1) 
    return (np.unique(x)) 

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')

df = pd.read_csv("combined_csv.csv")
prot_list = (list(df['prot_id']))
unique_list = unique(prot_list)
df_new = pd.DataFrame()
df_new['prot_id'] = unique_list

df_new.to_csv("rna_negative.csv",index=False,header= False)
```
